# SQL-Based Sales Data Visualization

## Introduction

This notebook presents a visual analysis of product sales data extracted from the Chinook SQLite database.

The objective is to transform structured SQL query outputs into meaningful visual insights using Python and Matplotlib. The analysis focuses on revenue distribution, time-based trends, and product-level performance.

The visualizations support business decision-making by highlighting sales patterns across regions and time periods.

```python
# ================================
# SALES DATA VISUALIZATION
# ================================

#matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset

# Ensure plots display clearly
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 12
plt.show()
```

Dataset Overview

The dataset was generated using SQL queries applied to the Chinook database.

The following outputs were exported as CSV files:

- Revenue by Country
- Monthly Revenue
- Top-Selling Tracks
- Ranked Tracks by Country (Window Function)

These structured outputs form the basis for visualization and business interpretation.

## Revenue by Country

This visualization represents total revenue generated per country.

Purpose:
- Identify high-performing markets
- Understand revenue concentration
- Support geographic sales strategy

```python
# ================================
# Revenue by Country
# ================================

# Load dataset
df_country = pd.read_csv("results/Country,Total_Revenue.csv")

# Sort by revenue
df_country = df_country.sort_values(by="Total_Revenue", ascending=False)

# Plot
plt.figure()
plt.bar(df_country["Country"], df_country["Total_Revenue"])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Total Revenue")
plt.title("Revenue by Country")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()

# Save figure
plt.savefig("screenshots/revenue_by_country.png")
plt.show()
```

## Monthly Revenue Trend

This time-series visualization illustrates revenue performance across months.

- Understand long-term revenue stability

## Top-Selling Tracks

This chart displays the top revenue-generating tracks.

Purpose:
- Identify high-demand products
- Understand product contribution to total revenue
- Support catalog optimization decisions

```python
# ================================
# Top 10 Tracks by Revenue
# ================================

df_tracks = pd.read_csv("results/top_selling_tracks.csv")

df_tracks = df_tracks.sort_values(by="Total_Revenue", ascending=True)

plt.figure()
plt.barh(df_tracks["TrackName"], df_tracks["Total_Revenue"])
plt.xlabel("Total Revenue")
plt.title("Top 10 Tracks by Revenue")
plt.tight_layout()

plt.savefig("screenshots/top_tracks.png")
plt.show()
```

## Key Insights

- Monthly trends reveal performance fluctuations, with identifiable peak periods.
- A limited number of tracks contribute significantly to total revenue.
